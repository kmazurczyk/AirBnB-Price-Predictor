```python
from __init__ import get_base_path
from dotenv import load_dotenv
from geopy.distance import geodesic
from google_api import get_address
from matplotlib import pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns
import sys

# environment
load_dotenv(get_base_path() + '/airbnb.env')
api_key = os.getenv('GOOGLE_API_KEY')
data_dir = get_base_path() + os.getenv('data_dir')

# Load Airbnb data
path = data_dir + 'processed/' + 'airbnb_cleansed_sample_with_landmark_features' + '.csv'
airbnb_data = pd.read_csv(path,index_col=0)
```

```python
airbnb_data['address'] = airbnb_data.apply(lambda df: get_address(df['latitude'], df['longitude'], api_key), axis=1)
```

```python
# Function to extract individual address components from Google API response get_address()
def get_address_component(json, component):

    result = [item['long_name'] for item in json['address_components'] if item['types'] == [component]]
    if len(result) > 0:
        return result[0]
    else:
        return None
```

```python
airbnb_data['zip_code'] = airbnb_data['cleaned'].apply(lambda x: get_address_component(x, 'postal_code'))
```
